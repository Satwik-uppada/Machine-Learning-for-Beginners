## <a id='toc1_2_'></a>[Loading Dataset](#toc0_)

```python
# Load the Wine dataset
wine_data = load_wine()
X = wine_data.data  # features
y = wine_data.target  # target
```

---

## <a id='toc1_3_'></a>[PCA](#toc0_)

...........................................................................................................................................................................................................................................................................................................................................................................................

### <a id='toc1_3_1_'></a>[Perform PCA](#toc0_)

```python
pca = PCA(n_components=2)  # reducing to 2 components
X_pca = pca.fit_transform(X)
```

...........................................................................................................................................................................................................................................................................................................................................................................................

### <a id='toc1_3_2_'></a>[Visualize the reduced dimensions](#toc0_)

```python
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA on Wine Dataset')
plt.colorbar()
plt.show()
```

...........................................................................................................................................................................................................................................................................................................................................................................................

### <a id='toc1_3_3_'></a>[significance](#toc0_)

```python
explained_variance = pca.explained_variance_ratio_
print("Explained variance by each component:", explained_variance)
print("Total explained variance:", np.sum(explained_variance))
```

## <a id='toc1_4_'></a>[Linear Discriminant Analysis (LDA)](#toc0_)


...........................................................................................................................................................................................................................................................................................................................................................................................

### <a id='toc1_4_1_'></a>[Perform LDA](#toc0_)

```python
lda = LDA(n_components=2)  # reducing to 2 components
X_lda = lda.fit_transform(X, y)
```

...........................................................................................................................................................................................................................................................................................................................................................................................

### <a id='toc1_4_2_'></a>[Visualize the reduced dimensions](#toc0_)

```python
plt.figure(figsize=(8,6))
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.title('LDA on Wine Dataset')
plt.colorbar()
plt.show()
```

...........................................................................................................................................................................................................................................................................................................................................................................................


### <a id='toc1_4_3_'></a>[Difference between PCA and LDA](#toc0_)
- **LDA aims to find a feature space that maximizes class separability whereas PCA aims to maximize variance.**


---

## <a id='toc1_5_'></a>[Non-Negative Matrix Factorization (NMF)](#toc0_)

### <a id='toc1_5_1_'></a>[Perform NMF](#toc0_)

```python
nmf = NMF(n_components=2, init='random', random_state=0)
X_nmf = nmf.fit_transform(X)
```


### <a id='toc1_5_2_'></a>[Visualize the reduced dimensions](#toc0_)

```python
plt.figure(figsize=(8,6))
plt.scatter(X_nmf[:, 0], X_nmf[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('NMF Component 1')
plt.ylabel('NMF Component 2')
plt.title('NMF on Wine Dataset')
plt.colorbar()
plt.show()
```

...........................................................................................................................................................................................................................................................................................................................................................................................


### <a id='toc1_5_3_'></a>[Interpretability of NMF components](#toc0_)
- **NMF components are often more interpretable because they are non-negative and can be seen as parts-based representation.**

## <a id='toc1_6_'></a>[T-Distributed Stochastic Neighbor Embedding (T-SNE)](#toc0_)

...........................................................................................................................................................................................................................................................................................................................................................................................


### <a id='toc1_6_1_'></a>[Perform t-SNE](#toc0_)

```python
tsne = TSNE(n_components=2, random_state=0)
X_tsne = tsne.fit_transform(X)
```

...........................................................................................................................................................................................................................................................................................................................................................................................

### <a id='toc1_6_2_'></a>[Visualize the reduced dimensions](#toc0_)

```python

plt.figure(figsize=(8,6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', edgecolor='k', s=50)
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE on Wine Dataset')
plt.colorbar()
plt.show()
```

...........................................................................................................................................................................................................................................................................................................................................................................................


### <a id='toc1_6_3_'></a>[Advantages and limitations of T-SNE](#toc0_)
- **T-SNE is effective for visualizing high-dimensional data in 2 or 3 dimensions. It preserves local structure but can distort global structure and is computationally expensive.**
